```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/dataset.csv')
data.head()
```

Data cleaning


```python
import re
import string
```

```python
for char in string.punctuation:
  print(char, end=' ')
translate_table=dict((ord(char),None)for char in string.punctuation)
```

```python
translate_table
```

```python
lines=data['languages']
data2=[]
for i in range(0,len(data)):
  line=lines[i]
  line=line.lower()
  line= re.sub(r"\d+", "", line)
  line= line.translate(translate_table)
  data2.append(line)
```

```python
data.shape
```

data split


```python
x,y = data2,data.iloc[:,2]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
xtrain
```

```python
import codecs
from sklearn import feature_extraction
from sklearn import linear_model
from sklearn import pipeline
from sklearn import metrics
```

Model

```python
vectorizer = feature_extraction.text.TfidfVectorizer(ngram_range=(1,3),analyzer='char')
pipe_lr_r13=pipeline.Pipeline([('vectorizer',vectorizer),('clf',linear_model.LogisticRegression())])
```

```python
model=pipe_lr_r13.fit(xtrain,ytrain)
```

```python
model
```

Accuracy

```python
y_pred=model.predict(xtest)
```

```python
acc=(metrics.accuracy_score(ytest,y_pred))*100
acc
```

Random example

```python
def lang_detect(tex):
  tex=" ".join(tex.split())
  tex=tex.lower()
  tex= re.sub(r"\d+", "", tex)
  tex= tex.translate(translate_table)
  predd=model.predict([tex])
  return predd[0]
```

```python
test="यह भाषा क्या है?"
lang_detect(test)
```

```python
import cv2
import easyocr
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 8,16
```

```python

reader= easyocr.Reader(['en','hi'])
```

```python
def img_lang_detect(image):
  img = cv2.imread(image)
  Text=reader.readtext(img)

  spe=[]
  for i in Text:
    spe.append(i[1])
  tex=" "
  tex=tex.join(spe)

  tex=" ".join(tex.split())
  tex=tex.lower()
  tex.strip()
  tex.replace('\n',' ')
  tex= re.sub(r"\d+", "", tex)
  tex= tex.translate(translate_table)
  print(tex)
  predd=model.predict([tex])
  return predd[0]
```

```python
example = "/content/download.jpg"
img_lang_detect(example)
```
